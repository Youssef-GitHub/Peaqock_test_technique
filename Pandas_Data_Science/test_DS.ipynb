### 13- Créer une fonction qui prend en paramètres une chaîne de caractères et qui transforme la dernière voyelle  en MAJUSCULE

```python
def convert_voy_maj(string):
    my_list = list(string[::-1])
    for i in range(len(my_list)):
        if my_list[i].upper() in ['A','E','I','O','U','Y']:
            my_list[i] = my_list[i].upper()
            break;
    return ''.join(my_list)[::-1]
```

### 14 - appliquer le fonction précedente pour transformer le dernièr caractère de la colonne 'Class' en majuscule

```python
new_df['Class'].apply(lambda x: convert_voy_maj(x))
```
